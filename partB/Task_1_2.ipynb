# Key Assumptions

### Assumption 1
* **Assumption**: The true probabilities of the data follow an s-shaped (sigmoid) curve when plotted against the best decision line.
* **Why the method needs it**: The method uses a specific reverse formula, $\sigma^{-1}(p) = -\log(1/p - 1)$, to create its targets. If the real data doesn't naturally fit this kind of s-shape, the targets will be wrong, and the model won't learn properly.
* **Violation scenario**: A dataset where the difference between classes is based on rare, extreme outliers rather than a smooth transition between groups.
* **Paper reference**: Section 3, "Inversion of Class Probability Estimation".

### Assumption 2
* **Assumption**: The average of the model's basic predictions for the items in a group is a good match for that group's transformed target value $y_i$.
* **Why the method needs it**: The formula tries to make the average prediction for a group, $\frac{1}{|S_i|} \sum_{j \in S_i} f(x_j)$, close to $y_i$. If this average doesn't truthfully represent the whole group, the model will learn the wrong pattern.
* **Violation scenario**: A group that has two very tight but separate clusters of points. The average might fall right in the empty middle, which doesn't represent either cluster well.
* **Paper reference**: Section 3, "Primal Problem" structural constraints equation $\frac{1}{|S_i|} \sum_{j \in S_i} (wx_j + b) \approx y_i$.

### Assumption 3
* **Assumption**: A simple linear estimate (first-order Taylor expansion) is good enough to calculate the error margins for the transformed targets.
* **Why the method needs it**: The method wants to keep the math easy to solve. So, it changes the basic error limit $\epsilon$ into group-specific limits using the formula $\epsilon_i = \epsilon / (p_i(1 - p_i))$. It ignores more complex math to keep things simple.
* **Violation scenario**: For groups where the probability is extremely close to 0 or 1, this simple formula creates massive error margins. This means the model would basically ignore these groups because any prediction would be considered "close enough."
* **Paper reference**: Section 3, the math steps under the primal problem showing how $\epsilon_i$ is calculated.
