#**Dataset Selection and Setup**

**Dataset Selection Justification**  
For this task, I chose the Breast Cancer Wisconsin (Diagnostic) dataset from `scikit-learn` (569 samples, 30 features).  
It is a reasonable testbed for the Inverse Calibration method because it is a binary classification problem where we can randomly partition the training set into groups of size k (as done in Section 4 of the paper) and compute the group label frequencies pk exactly as required by the algorithm.  
The dataset size and number of features are comparable to the UCI datasets used in the original experiments (e.g., breast-w has 699 samples and 10 features).  
The only limitation is that the original paper also tested multi-class datasets such as primary-tumor and soybean, while this dataset is strictly binary.  
I preprocessed the data by standard-scaling the features, converting labels to {-1, +1}, and splitting into 70/30 train/test sets for later reproduction.

```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import os

# Set a random seed so results are the same every time
np.random.seed(42)

os.makedirs('partB/data', exist_ok=True)
data = load_breast_cancer()

# We scale the features so the SVM models work better
X = StandardScaler().fit_transform(data.data)

# The paper's math needs the labels to be -1 and 1
y = np.where(data.target == 0, -1, 1)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Save the prepared data so other notebooks can use it
np.save('partB/data/X_train.npy', X_train)
np.save('partB/data/X_test.npy', X_test)
np.save('partB/data/y_train.npy', y_train)
np.save('partB/data/y_test.npy', y_test)

print(f"Data prepared: X_train shape {X_train.shape}, y_train shape {y_train.shape}")

```
